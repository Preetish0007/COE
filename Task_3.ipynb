```python

def reverse_words(sentence):

    words = sentence.split()

    reversed_words = words[::-1]
    # Join the reversed list of words into a single string
    reversed_sentence = ' '.join(reversed_words)
    return reversed_sentence
sentence = "My name is Preetish "
reversed_sentence = reverse_words(sentence)
print(reversed_sentence)
```

```python
import pandas as pd
data=pd.read_csv("/content/StudentsPerformance.csv")
```

# New Section

```python
# Assuming 'data' is the DataFrame containing the student performance data
data['math_pass'] = ['Pass' if score >= 33 else 'Fail' for score in data['math score']]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
data['math_pass'].value_counts().plot(kind='bar')
plt.title('Comparison of students passed or failed in maths')
plt.xlabel('status')
plt.ylabel('count')
plt.show()
```
